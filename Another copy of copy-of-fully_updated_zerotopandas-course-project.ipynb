Handle missing, incorrect and invalid data

````python
# Fill missing values for numerical columns with the mean
for col in shopping_df.select_dtypes(include = "number").columns:
    if shopping_df[col].isnull().sum() > 0 :
        shopping_df[col].fillna(shopping_df[col].mean(), inplace = True)
````

````python
# filling the missing values  for categorical columns with the mode

for col in shopping_df.select_dtypes(include ="object").columns:
    if shopping_df[col].isnull().sum() > 0:
        shopping_df[col].fillna(shopping_df[col].mode(), inplace = True)
````

This checks all numerical columns and converts invalid values to NaN e.g., if a number column has "abc", it will become NaN

An invalid numeric value is any value that cannot be interpreted as a number. Examples include:

Strings: "abc", "N/A", "unknown".
Special characters: "@!", "#$$".
Dates or mixed data: "01/01/2023", "45abc".
Without errors='coerce', these values would cause an error when trying to convert them to numbers.

````python
for col in shopping_df.select_dtypes(include ="number").columns:
    shopping_df[col] = pd.to_numeric(shopping_df[col],errors = "coerce")
````

Fix Invalid Data in Categorical Columns:

Replace invalid or placeholder values like empty strings ('') or "Unknown" with NaN.

````python
for col in shopping_df.select_dtypes(include ="object").columns:
    shopping_df[col] = shopping_df[col].replace([" ","Unknown"],pd.NA,inplace =True)
````

Purchase Frequency Heatmap by Payment Method and Shipping Type
Question: "Which combinations of payment methods and shipping types are most common?"

````python
# Heatmap
heatmap_data = shopping_df.pivot_table(
    values='Customer ID',
    index='Payment Method',
    columns='Shipping Type',
    aggfunc='count'
)

plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=True, cmap='Oranges', fmt=".0f", linewidths=0.5)
plt.title('Purchase Frequency by Payment Method and Shipping Type', fontsize=16)
plt.xlabel('Shipping Type', fontsize=14)
plt.ylabel('Payment Method', fontsize=14)
plt.tight_layout()
plt.show()
````

How does the total number of purchases vary across age groups?"


````python
shopping_df['Age Group'] = pd.cut(
    shopping_df['Age'], bins=[0, 18, 35, 50, 65, 100], labels=['<18', '18-35', '36-50', '51-65', '65+']
)

total_purchases_age_group = shopping_df.groupby('Age Group')['Customer ID'].count()

plt.figure(figsize=(8, 6))
total_purchases_age_group.plot(marker='s', linestyle='--', color='green')
plt.title('Total Purchases by Age Group', fontsize=16)
plt.xlabel('Age Group', fontsize=14)
plt.ylabel('Total Purchases', fontsize=14)
plt.grid(alpha=0.5)
plt.tight_layout()
plt.show()
````

  What is the proportion of purchases over $50 across different subscription statuses?

````python
import pandas as pd
import matplotlib.pyplot as plt

# Group by 'Subscription Status' and calculate the proportion of purchases over $50
purchase_over_50_proportion = shopping_df.groupby('Subscription Status')['purchase_over_fifty'].mean()

# Plot the results
plt.figure(figsize=(10, 6))
bars = purchase_over_50_proportion.plot(kind='bar', color=['cornflowerblue', 'salmon'], edgecolor='black', alpha=0.9)

# Add annotations for the bar values
for i, value in enumerate(purchase_over_50_proportion):
    plt.text(i, value + 0.02, f'{value:.2%}', ha='center', va='bottom', fontsize=10, color='black')

# Add descriptive labels and titles
plt.title('Proportion of Purchases Over $50 by Subscription Status', fontsize=16)
plt.xlabel('Subscription Status', fontsize=14)
plt.ylabel('Proportion of Purchases Over $50', fontsize=14)
plt.xticks(rotation=0, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
````

How do review ratings differ across product categories and genders

````python
heatmap_data = shopping_df.pivot_table(
    values='Review Rating',
    index='Category',
    columns='Gender',
    aggfunc='mean'
)

plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=True, fmt=".2f", cmap='Blues', linewidths=0.5)
plt.title('Average Review Rating by Product Category and Gender', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Product Category', fontsize=14)
plt.tight_layout()
plt.show()
````

A line chart provides a clear view of how average spending changes across seasons, separated by gender.

````python
import matplotlib.pyplot as plt

# Group data by Season and Gender, calculate average purchase amount
avg_purchase_season_gender = shopping_df.groupby(['Season', 'Gender'])['Purchase Amount (USD)'].mean().unstack()

# Plotting the line chart
plt.figure(figsize=(10, 6))
for gender in avg_purchase_season_gender.columns:
    plt.plot(avg_purchase_season_gender.index, avg_purchase_season_gender[gender], marker='o', label=gender)

# Add titles and labels
plt.title('Average Purchase Amount by Season and Gender', fontsize=16)
plt.xlabel('Season', fontsize=14)
plt.ylabel('Average Purchase Amount (USD)', fontsize=14)
plt.legend(title="Gender", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
````

WHAT IS THE AVERAGE PURCHASE FOR BLOUSES PER WEEK

````python
shopping_df["Frequency of Purchases"] = pd.to_numeric(shopping_df["Frequency of Purchases"], errors="coerce")
shopping_df["Purchase Amount (USD)"] = pd.to_numeric(shopping_df["Purchase Amount (USD)"], errors="coerce")

# Step 2: Filter for 'Blouse' purchases
blouse_purchases = shopping_df[shopping_df["Item Purchased"] == "Blouse"]

# Step 3: Calculate weekly average only for 'Blouse' rows
shopping_df["Weekly_Average"] = None  # Initialize column
shopping_df.loc[shopping_df["Item Purchased"] == "Blouse", "Weekly_Average"] = (
    blouse_purchases["Purchase Amount (USD)"] / blouse_purchases["Frequency of Purchases"]
)

# Display the updated DataFrame
print(shopping_df)





````

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(18, 10))

# Boxplot: Distribution of purchase amount by gender
sns.boxplot(x="Gender", y="Purchase Amount (USD)", data=shopping_df, palette="coolwarm", ax=axes[0, 0])
axes[0, 0].set_title("Distribution of Purchase Amount by Gender", fontsize=14)
axes[0, 0].set_xlabel("Gender", fontsize=12)
axes[0, 0].set_ylabel("Purchase Amount (USD)", fontsize=12)

# Countplot: Count of purchases by gender and subscription status
sns.countplot(x="Gender", hue="Subscription Status", data=shopping_df, palette="Set2", ax=axes[0, 1])
axes[0, 1].set_title("Count of Purchases by Gender and Subscription Status", fontsize=14)
axes[0, 1].set_xlabel("Gender", fontsize=12)
axes[0, 1].set_ylabel("Count of Purchases", fontsize=12)
axes[0, 1].legend(title="Subscription Status", loc="upper right")

# Lineplot: Average purchase amount by age
age_avg_purchase = shopping_df.groupby("Age")["Purchase Amount (USD)"].mean().reset_index()
sns.lineplot(x="Age", y="Purchase Amount (USD)", data=age_avg_purchase, marker="o", color="purple", ax=axes[1, 0])
axes[1, 0].set_title("Average Purchase Amount by Age", fontsize=14)
axes[1, 0].set_xlabel("Age", fontsize=12)
axes[1, 0].set_ylabel("Average Purchase Amount (USD)", fontsize=12)

# Stacked Bar Plot: Purchase frequency by location and category
location_category_counts = shopping_df.groupby(["Location", "Category"]).size().unstack(fill_value=0)
location_category_counts.plot(kind="bar", stacked=True, colormap="viridis", ax=axes[1, 1])
axes[1, 1].set_title("Purchase Frequency by Location and Category", fontsize=14)
axes[1, 1].set_xlabel("Location", fontsize=12)
axes[1, 1].set_ylabel("Purchase Frequency", fontsize=12)
axes[1, 1].legend(title="Category", fontsize=10)

# Adjust layout for better spacing
plt.tight_layout(pad=4)
plt.show()
````

````python
import pandas as pd

# Dataset 1: Customer Purchases
shop_data = {
    "Customer ID": ["C001", "C002", "C003", "C004", "C005"],
    "Item Purchased": ["Blouse", "Shirt", "Pants", "Skirt", "Dress"],
    "Purchase Amount (USD)": [200, 50, 300, 150, 400],
    "Store": ["Store A", "Store B", "Store A", "Store C", "Store D"]
}
shop_df = pd.DataFrame(shop_data)

# Dataset 2: Store Details
store_info_data = {
    "Store": ["Store A", "Store B", "Store C", "Store D"],
    "Location": ["New York", "Los Angeles", "Chicago", "Houston"],
    "Manager": ["Alice", "Bob", "Carol", "David"]
}
store_info_df = pd.DataFrame(store_info_data)

# Merge the two datasets on 'Store'
merged_df = pd.merge(shop_df, store_info_df, on="Store")

# Drop the 'Location' column
merged_df = merged_df.drop(columns=["Location"])

# Calculate total purchase amount per store
store_totals = merged_df.groupby(["Store", "Manager"])["Purchase Amount (USD)"].sum()

# Find the store with the highest total purchase amount
highest_store = store_totals.idxmax()
highest_amount = store_totals.max()

print("Merged Dataset without 'Location':")
print(merged_df)
print("\nTotal Purchase Amount by Store:")
print(store_totals)
print(f"\nStore with the highest purchase amount: {highest_store}, Amount: ${highest_amount}")
````
